## Buscando os dados 

Site: http://dados.cvm.gov.br/dataset/fi-doc-inf_diario/resource/85eb3cea-a071-4d3c-9ace-0fcdb6807e14

```python
def busca_fundos(ano, mes):
  url = 'http://dados.cvm.gov.br/dados/FI/DOC/INF_DIARIO/DADOS/inf_diario_fi_{:02d}{:02d}.csv'.format(ano, mes)
  dados = pd.read_csv(url, sep=';')
  return dados
```

```python
fundos12 = busca_fundos(2022, 1)
fundos12.head()
```

##Filtro por cotistas

```python
fundos12.sort_values(['NR_COTST'], ascending=False, inplace=True)
fundos12.head()
```

```python
estatistica_cotistas = pd.DataFrame(fundos12['NR_COTST'].describe())
estatistica_cotistas
```

```python
#Determinando a quantidade de cotistas.

cotistas = 100000

fundos_filtro = fundos12[fundos12['NR_COTST'] >= cotistas]
print(len(fundos_filtro))
```

##Melhores e piores 5 fundos

```python
#Problema: cotação misturada

melhores = pd.DataFrame()
melhores["CNPJ"] = fundos_filtro['CNPJ_FUNDO']
melhores['Cota'] = fundos_filtro['VL_QUOTA']
melhores.head()
```
