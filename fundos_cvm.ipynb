## Bibliotecas

## Buscando os dados 

Site: http://dados.cvm.gov.br/dataset/fi-doc-inf_diario/resource/85eb3cea-a071-4d3c-9ace-0fcdb6807e14

```python
def busca_fundos(ano, mes):
  url = 'http://dados.cvm.gov.br/dados/FI/DOC/INF_DIARIO/DADOS/inf_diario_fi_{:02d}{:02d}.csv'.format(ano, mes)
  dados = pd.read_csv(url, sep=';')
  return dados
```

```python
fundos_mes = busca_fundos(2021, 12)
fundos_mes.head()
```

```python
fundos_mes.sort_values(['NR_COTST'], ascending=False, inplace=True)
fundos_mes.head()
```

```python
estatistica_cotistas = pd.DataFrame(fundos_mes['NR_COTST'].describe())
estatistica_cotistas
```

```python
#Determinando a quantidade de cotistas.

cotistas = 1000
```

```python
fundos_filtro = fundos_mes[fundos_mes['NR_COTST'] >= cotistas].pivot(index='DT_COMPTC', columns='CNPJ_FUNDO', 
                                                                     values=['VL_TOTAL', 'VL_QUOTA', 'VL_PATRIM_LIQ', 'CAPTC_DIA', 'RESG_DIA'])
print(len(fundos_filtro))
```

```python
fundos_filtro.head()
```

```python
#Normalizando a base

normalizacao = fundos_filtro['VL_QUOTA'] / fundos_filtro['VL_QUOTA'].iloc[0]
normalizacao.head()
```

```python
#Retorno simples
retorno_simples = np.log(fundos_filtro['VL_QUOTA'] / fundos_filtro['VL_QUOTA'].shift())

#Retorno acumulado
retorno_acumulado = (1 + retorno_simples).cumprod()
retorno_acumulado.iloc[0] = 1

retorno_acumulado.tail()
```

## Buscando nome dos fundos

```python
#Nome dos fundos 

url1 = 'http://dados.cvm.gov.br/dados/FI/CAD/DADOS/cad_fi.csv'
```

```python
nomes = pd.read_csv(url1, sep=';', encoding='ISO-8859-1')
```

```python
nomes.tail()
```

## Melhores e piores fundos

```python
#5 melhores fundos
melhores_fundos = pd.DataFrame()
melhores_fundos['Retorno (%)'] = (normalizacao.iloc[-1].sort_values(ascending=False)[:5] - 1) * 100
melhores_fundos
```

```python
for cnpj in melhores_fundos.index:
    fundo = nomes[nomes['CNPJ_FUNDO'] == cnpj]
    melhores_fundos.at[cnpj, 'Nome do fundo'] = fundo['DENOM_SOCIAL'].values[0]
    melhores_fundos.at[cnpj, 'Classe'] = fundo['CLASSE'].values[0]
    melhores_fundos.at[cnpj, 'PL'] = fundo['VL_PATRIM_LIQ'].values[0]
    
melhores_fundos
```

```python
#5 piores fundos

piores_fundos = pd.DataFrame()
piores_fundos['Retorno (%)'] = (normalizacao.iloc[-1].sort_values(ascending=True)[:5] - 1) * 100
piores_fundos
```

```python
for cnpj in piores_fundos.index:
    fundo1 = nomes[nomes['CNPJ_FUNDO'] == cnpj]
    piores_fundos.at[cnpj, 'Nome do fundo'] = fundo1['DENOM_SOCIAL'].values[0]
    piores_fundos.at[cnpj, 'Classe'] = fundo1['CLASSE'].values[0]
    piores_fundos.at[cnpj, 'PL'] = fundo1['VL_PATRIM_LIQ'].values[0]
    
piores_fundos
```
